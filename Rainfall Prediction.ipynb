```python
# Handling missing values in numerical columns (float64)
numerical_columns = ['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine', 'WindGustSpeed',
                     'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm', 'Pressure9am',
                     'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm']

for column in numerical_columns:
    # Impute missing values with the column mean
    df[column].fillna(df[column].mean(), inplace=True)

# Handling missing values in categorical columns (object)
categorical_columns = ['Date', 'Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm', 'RainToday', 'RainTomorrow']

for column in categorical_columns:
    # Impute missing values with the most frequent category
    df[column].fillna(df[column].mode()[0], inplace=True)

# Verify that there are no missing values remaining
missing_values_after = df.isnull().sum()
print(missing_values_after)

```

```python


# Exclude datetime columns
numeric_columns = df.select_dtypes(include=[np.number]).columns

# Detect outliers using IQR
Q1 = df[numeric_columns].quantile(0.25)
Q3 = df[numeric_columns].quantile(0.75)
IQR = Q3 - Q1
outlier_threshold = 1.5  # Adjust this threshold as per your requirement

# Iterate over each column and detect outliers
for column in numeric_columns:
    lower_bound = Q1[column] - outlier_threshold * IQR[column]
    upper_bound = Q3[column] + outlier_threshold * IQR[column]
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"Column: {column}")
    print("Number of outliers:", len(outliers))
    print("Outlier rows:", outliers)
    print()

```

```python
# Define outlier threshold
outlier_threshold = 3  # Adjust this threshold as per your requirement

# Iterate over each column and treat outliers
for column in df.columns:
    if df[column].dtype in [np.float64, np.int64]:
        # Calculate the lower and upper bounds for outliers
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - outlier_threshold * IQR
        upper_bound = Q3 + outlier_threshold * IQR

        # Cap or threshold the outliers
        df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
        df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])

# Updated dataset with treated outliers
print(df.head())
```

```python
#Chekcing Skewness

# Select numerical columns
numeric_columns = df.select_dtypes(include=[np.number]).columns

# Calculate skewness of each column
skewness = df[numeric_columns].skew()

# Plot histogram and density plots for each column
for column in numeric_columns:
    fig, ax = plt.subplots(1, 2, figsize=(10, 4))
    
    # Histogram plot
    sns.histplot(data=df, x=column, kde=False, ax=ax[0])
    ax[0].set_title(f"Histogram of {column}")
    
    # Density plot
    sns.kdeplot(data=df, x=column, ax=ax[1])
    ax[1].set_title(f"Density Plot of {column}")
    
    plt.tight_layout()
    plt.show()

    # Print skewness
    print(f"Skewness of {column}: {skewness[column]}")
    print()
```

# Balancing the target variable



```python
from sklearn.utils import resample
# Separate the majority and minority classes
majority_class = df[df['RainTomorrow'] == 0]
minority_class = df[df['RainTomorrow'] == 1]

# Upsample the minority class (oversampling)
minority_upsampled = resample(minority_class,
                              replace=True,
                              n_samples=len(majority_class),
                              random_state=42)

# Downsample the majority class (undersampling)
majority_downsampled = resample(majority_class,
                                replace=False,
                                n_samples=len(minority_class),
                                random_state=42)

# Combine the upsampled minority class with the majority class
upsampled_data = pd.concat([majority_class, minority_upsampled])

# Combine the downsampled majority class with the minority class
downsampled_data = pd.concat([majority_downsampled, minority_class])

# Print the class distribution after balancing
print("Upsampled Data - Class Distribution:")
print(upsampled_data['RainTomorrow'].value_counts())

print("\nDownsampled Data - Class Distribution:")
print(downsampled_data['RainTomorrow'].value_counts())
```

ML Model

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder

# Convert categorical variables using one-hot encoding
categorical_columns = ['categorical_columns']  # Replace with the actual name of the categorical column(s)
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
encoded_categorical = pd.DataFrame(encoder.fit_transform(df[categorical_columns]))
encoded_categorical.columns = encoder.get_feature_names(categorical_columns)
df.drop(categorical_columns, axis=1, inplace=True)
df = pd.concat([df, encoded_categorical], axis=1)

# Split the dataset into features (X) and target variables (y)
X = df.drop("RainTomorrow", axis=1)  # Features
y_classification = df["RainTomorrow"]  # Binary classification target variable

# Split the dataset into training and testing sets
X_train, X_test, y_train_classification, y_test_classification = train_test_split(X, y_classification, test_size=0.2, random_state=42)

# Classification Models
classification_models = [
    LogisticRegression(),
    RandomForestClassifier(),
    GradientBoostingClassifier(),
    SVC(),
    KNeighborsClassifier(),
    DecisionTreeClassifier(),
    AdaBoostClassifier()
]

# Train and evaluate classification models
for model in classification_models:
    model.fit(X_train, y_train_classification)
    predictions = model.predict(X_test)
    accuracy = accuracy_score(y_test_classification, predictions)
    print(f"Classification Model: {type(model).__name__}")
    print("Accuracy:", accuracy)
    print()

```
